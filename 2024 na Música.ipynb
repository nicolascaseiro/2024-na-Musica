```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/nicolascaseiro/2024-na-Musica/refs/heads/main/2024%20na%20M%C3%BAsica.csv", encoding='utf-8')

df.head()
```

```python
colunas_numericas = df.select_dtypes(include=['number']).columns

outliers_bool = pd.DataFrame(False, index=df.index, columns=colunas_numericas)

for col in colunas_numericas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers_bool[col] = (df[col] < limite_inferior) | (df[col] > limite_superior)

linhas_com_outliers = outliers_bool.any(axis=1)

df_outliers = df[linhas_com_outliers]

print(f"Total de linhas com pelo menos um outlier: {df_outliers.shape[0]}")
```

```python
def tempo_para_segundos(tempo_str):
    try:
        minutos, segundos = tempo_str.split(':')
        return int(minutos) * 60 + int(segundos)
    except:
        return None

df['Duração_segundos'] = df['Duração'].apply(tempo_para_segundos)

df['Duração_segundos'] = pd.to_numeric(df['Duração_segundos'], errors='coerce')

duracao_numerica = df['Duração_segundos'].dropna()
Q1 = duracao_numerica.quantile(0.25)
Q3 = duracao_numerica.quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df['Duração_segundos'] < limite_inferior) | (df['Duração_segundos'] > limite_superior)]

print(f"Número de outliers na duração: {outliers.shape[0]}\n")

print(outliers[['Música', 'Artista', 'Duração']])
```

```python
colunas = list(df.columns)

colunas.remove('Duração_segundos')

idx = colunas.index('Duração')

colunas.insert(idx + 1, 'Duração_segundos')

df = df[colunas]

media_duracao = df['Duração_segundos'].mean()
minutos = int(media_duracao // 60)
segundos = int(media_duracao % 60)
print(f"Média da duração das músicas: {minutos} min {segundos} seg")
```

```python
df['BPM'] = pd.to_numeric(df['BPM'], errors='coerce')

bpm_validos = df['BPM'].dropna()

Q1 = bpm_validos.quantile(0.25)
Q3 = bpm_validos.quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers_bpm = df[(df['BPM'] < limite_inferior) | (df['BPM'] > limite_superior)]

print(f"Número de outliers de BPM: {outliers_bpm.shape[0]}")
display(outliers_bpm[['Música', 'Artista', 'BPM']])
```

```python
media_bpm = df['BPM'].mean()
print(f"Média do BPM: {media_bpm:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

generos_temp = df['Gêneros'].dropna().apply(lambda x: [g.strip() for g in x.split(',')])
df_temp = df.loc[generos_temp.index].copy()
df_temp['genero'] = generos_temp
df_temp = df_temp.explode('genero')

contagem_generos = df_temp['genero'].value_counts()

plt.figure(figsize=(10,6))
contagem_generos.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Gêneros Musicais')
plt.ylabel('Quantidade de Músicas')
plt.xlabel('Gêneros')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as ticker

def formatar_tempo(segundos, pos=None):
    minutos = int(segundos) // 60
    seg = int(segundos) % 60
    return f'{minutos}:{seg:02d}'

plt.figure(figsize=(10, 6))
plt.hist(df['Duração_segundos'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribuição da Duração das Músicas')
plt.xlabel('Duração (min:seg)')
plt.ylabel('Número de Músicas')

ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(formatar_tempo))

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(df['Duração_segundos'].dropna(), vert=False)
plt.title('Boxplot da Duração das Músicas')
plt.xlabel('Duração (min:seg)')

ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(formatar_tempo))

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Dashboard 2024 na Música", layout="wide")

def carregar_dados():
    url = "https://raw.githubusercontent.com/nicolascaseiro/2024-na-Musica/refs/heads/main/2024%20na%20M%C3%BAsica.csv"
    df = pd.read_csv(url, encoding='utf-8')
    return df

df = carregar_dados()

generos_lista = df['Gêneros'].dropna().apply(lambda x: [g.strip() for g in x.split(',')])
artistas_lista = df['Artista'].dropna().apply(lambda x: [a.strip() for a in x.split(',')])

df_temp = df.copy()
df_temp = df_temp.loc[generos_lista.index]
df_temp = df_temp.assign(Gêneros_lista=generos_lista)
df_temp = df_temp.explode('Gêneros_lista')

artistas_lista_temp = artistas_lista.loc[df_temp.index]
df_temp = df_temp.assign(Artistas_lista=artistas_lista_temp)
df_temp = df_temp.explode('Artistas_lista')

meses_pt = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
            'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

df_temp['Data do Álbum'] = pd.to_datetime(df_temp['Data do Álbum'], errors='coerce')
df_temp['Mês'] = df_temp['Data do Álbum'].dt.month.apply(lambda x: meses_pt[x-1] if pd.notna(x) else None)

st.sidebar.header("Filtros")

def ordenar_meses(meses_selecionados):
    return sorted(meses_selecionados, key=lambda m: meses_pt.index(m))

meses_disponiveis = ordenar_meses(df_temp['Mês'].dropna().unique())
generos_disponiveis = sorted(df_temp['Gêneros_lista'].dropna().unique())
artistas_disponiveis = sorted(df_temp['Artistas_lista'].dropna().unique())

mes_selecionado = st.sidebar.multiselect('Filtrar por Mês:', meses_disponiveis)
genero_selecionado = st.sidebar.multiselect('Filtrar por Gênero:', generos_disponiveis)
artista_selecionado = st.sidebar.multiselect('Filtrar por Artista:', artistas_disponiveis)

df_filtrado = df_temp.copy()

if mes_selecionado:
    df_filtrado = df_filtrado[df_filtrado['Mês'].isin(mes_selecionado)]

if genero_selecionado:
    df_filtrado = df_filtrado[df_filtrado['Gêneros_lista'].isin(genero_selecionado)]

if artista_selecionado:
    df_filtrado = df_filtrado[df_filtrado['Artistas_lista'].isin(artista_selecionado)]

indices_filtrados = df_filtrado.index.unique()
df_tabela = df.loc[indices_filtrados].copy()
df_tabela['Data do Álbum'] = pd.to_datetime(df_tabela['Data do Álbum'], errors='coerce')
df_tabela['Mês'] = df_tabela['Data do Álbum'].dt.month.apply(lambda x: meses_pt[x-1] if pd.notna(x) else None)
total_musicas_filtradas = df_tabela['Música'].nunique()

st.title("Dashboard 2024 na Música")

col1, col2, col3, col4 = st.columns(4)
col1.metric("Total de músicas", total_musicas_filtradas)
col2.metric("Média Popularidade", f"{df_filtrado['Popularidade'].mean():.2f}")
col3.metric("Total de Artistas", df_filtrado['Artistas_lista'].nunique())
col4.metric("Total de Gêneros", df_filtrado['Gêneros_lista'].nunique())

st.markdown("---")

df_grafico = df_filtrado.groupby('Gêneros_lista')['Popularidade'].mean().reset_index()

fig = px.bar(df_grafico,
             x='Gêneros_lista',
             y='Popularidade',
             title='Popularidade Média por Gênero Musical (2024)',
             labels={'Gêneros_lista': 'Gênero', 'Popularidade': 'Popularidade Média'},
             color='Popularidade',
             color_continuous_scale='Viridis',
             text=df_grafico['Popularidade'].round(2)
            )

fig.update_traces(textposition='outside', marker_line_width=1.5, marker_line_color='black')
fig.update_layout(
    plot_bgcolor='white',
    title_font=dict(size=24, family='Verdana', color='darkblue'),
    xaxis_title_font=dict(size=18, family='Verdana'),
    yaxis_title_font=dict(size=18, family='Verdana'),
    xaxis_tickangle=-45,
    xaxis_tickfont=dict(size=12, family='Verdana'),
    yaxis=dict(range=[0, 100]),
    margin=dict(l=40, r=40, t=80, b=100)
)

st.plotly_chart(fig, use_container_width=True)

st.markdown("---")

colunas_exibir = ['Música', 'Artista', 'Gêneros', 'Popularidade', 'Mês']
st.dataframe(df_tabela[colunas_exibir])
```
